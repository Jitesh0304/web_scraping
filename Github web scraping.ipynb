```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url = "https://github.com/topics"
```

```python
responce = requests.get(url) 
```

```python
responce.status_code        
```

```python
len(responce.text)     
```

```python
page_contents = responce.text
```

```python
page_contents[:100]
```

```python
soup = BeautifulSoup(page_contents , 'html.parser') 
```

```python
type(soup)
```

```python
topic_title_tags = soup.find_all('p',{'class': 'f3 lh-condensed mb-0 mt-1 Link--primary'})
```

```python
len(topic_title_tags)
```

```python
description ="f5 color-fg-muted mb-0 mt-1"

topic_desc_tags = soup.find_all('p',class_ = description) 
```

```python
len(topic_desc_tags)
```

```python
topic_title_tags[0]
```

```python
topic_link_tags = soup.find_all('a', class_ ='no-underline flex-1 d-flex flex-column')
```

```python
len(topic_link_tags)
```

```python
topic_link_tags[0]
```

```python
topic_link_tags[0]['href']     
```

```python
topic_titles = []     
for tag in topic_title_tags:           
    topic_titles.append(tag.text)   
                                     
topic_titles
```

```python
topic_description = []

for tag in topic_desc_tags:
    topic_description.append(tag.text)
    
topic_description
```

```python
topic_description = []

for tag in topic_desc_tags:
    topic_description.append(tag.text.strip())
    
topic_description
```

```python
topic_urls = []

for urrls in topic_link_tags:
    topic_urls.append(urrls['href'])
topic_urls
```

```python
topic_urls = []
base_url = 'https://github.com'

for urrls in topic_link_tags:
    topic_urls.append(base_url + urrls['href'])
    
topic_urls
```

```python
len(topic_urls)
```

```python
documents = {'Title':topic_titles,
            'Description':topic_description,
            'Url_Links':topic_urls}

datframe = pd.DataFrame(documents)

datframe
```

```python
datframe.to_csv('Github_data.csv')
```
